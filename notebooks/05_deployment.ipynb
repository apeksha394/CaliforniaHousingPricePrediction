```python
import joblib

model = joblib.load(
    "../models/best_tuned_model.pkl"
)
```

```python
preprocessor = joblib.load(
    "../models/preprocessor.pkl"
)
```

```python
sample_data = {
    "longitude": -122.23,
    "latitude": 37.88,
    "housing_median_age": 41,
    "total_rooms": 880,
    "total_bedrooms": 129,
    "population": 322,
    "households": 126,
    "median_income": 8.3252,
    "ocean_proximity": "NEAR OCEAN"
}
```

```python
import pandas as pd

sample_df = pd.DataFrame([sample_data])
```

```python
sample_df["rooms_per_household"] = (
    sample_df["total_rooms"] /
    sample_df["households"]
)

sample_df["bedrooms_per_room"] = (
    sample_df["total_bedrooms"] /
    sample_df["total_rooms"]
)

sample_df["population_per_household"] = (
    sample_df["population"] /
    sample_df["households"]
)
```

```python
processed_input = preprocessor.transform(
    sample_df
)
```

```python
prediction = model.predict(
    processed_input
)
```
